```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
com_school_data_df = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
com_school_data_df.columns
```

```python
total_number_of_schools = len(com_school_data_df["school_name"].unique())

total_number_of_students = len(com_school_data_df["Student ID"].unique())

total_budget = school_data["budget"].sum()

average_math_score = round(com_school_data_df["math_score"].mean(),2)

average_reading_score = round(com_school_data_df["reading_score"].mean(),2)

#These dataframes are used to calculate passing percentages
passing_math_df = com_school_data_df.loc[com_school_data_df["math_score"] >= 70,:]
passing_reading_df = com_school_data_df.loc[com_school_data_df["reading_score"] >= 70,:]
total_passing_df = pd.merge(passing_math_df,passing_reading_df,how="inner")

passing_math_perc = round((len(passing_math_df["student_name"].unique())/total_number_of_students)*100,2)

passing_reading_perc = round((len(passing_reading_df["student_name"].unique())/total_number_of_students)*100,2)

total_passing_perc = round((len(total_passing_df["student_name"].unique())/total_number_of_students)*100,2)

district_summary_df = pd.DataFrame({
    "Total Number of Schools": [total_number_of_schools],
    "Total Number of Students": [total_number_of_students],
    "Total School Budget":[total_budget],
    "Average Math Score":[average_math_score],
    "Average Reading Score":[average_reading_score],
    "Students Passing Math (percent)":[passing_math_perc],
    "Students Passing Reading (percent)":[passing_reading_perc],
    "Students Passing Overall (percent)":[total_passing_perc]
    })
district_summary_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
school_groups = com_school_data_df.groupby(["school_name"])

#ss prefix signifies school specific variables
ss_total_students = school_groups["student_name"].count()
ss_average_math_score = school_groups["math_score"].mean()
ss_average_reading_score = school_groups["reading_score"].mean()
ss_school_budget = school_groups["budget"].first()
ss_budget_per_student = ss_school_budget/ss_total_students
ss_school_types = school_groups["type"].first()

#using previous dataframes to get school specific passing stats
ss_passing_math_group = passing_math_df.groupby(["school_name"])
ss_passing_reading_group = passing_reading_df.groupby(["school_name"])
ss_passing_overall_group = total_passing_df.groupby(["school_name"])

#calculating percentages of students passing subjects
ss_passing_reading = round(ss_passing_reading_group["student_name"].count()/ss_total_students*100,2)
ss_passing_math = round(ss_passing_math_group["student_name"].count()/ss_total_students*100,2)
ss_passing_overall = round(ss_passing_overall_group["student_name"].count()/ss_total_students*100,2)

school_group_df = pd.DataFrame({
    "School Type":ss_school_types,
    "Total Students":ss_total_students,
    "Total School Budget":ss_school_budget,
    "Budget Per Student":ss_budget_per_student,
    "Average Math Score":ss_average_math_score,
    "Average Reading Score":ss_average_reading_score,
    "% Passing Math":ss_passing_math,
    "% Passing Reading":ss_passing_reading,
    "% Passing Overall":ss_passing_overall
})

school_group_df
```

## Top Performing Schools (By % Overall Passing)

```python
top_schools_df = school_group_df.sort_values("% Passing Overall",ascending=False)
top_schools_df.head()
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
#create series for each grade level
ninth_grade = com_school_data_df.loc[com_school_data_df["grade"] == "9th",:]
tenth_grade = com_school_data_df.loc[com_school_data_df["grade"] == "10th",:]
eleventh_grade = com_school_data_df.loc[com_school_data_df["grade"] == "11th",:]
twelfth_grade = com_school_data_df.loc[com_school_data_df["grade"] == "12th",:]

#group series by school
ninth_grade_by_school = ninth_grade.groupby(["school_name"])
tenth_grade_by_school = tenth_grade.groupby(["school_name"])
eleventh_grade_by_school = eleventh_grade.groupby(["school_name"])
twelfth_grade_by_school = twelfth_grade.groupby(["school_name"])

math_scores_by_grade_df = pd.DataFrame({
    
    "9th":ninth_grade_by_school["math_score"].mean(),
    "10th":tenth_grade_by_school["math_score"].mean(),
    "11th":eleventh_grade_by_school["math_score"].mean(),
    "12th":twelfth_grade_by_school["math_score"].mean()
})
math_scores_by_grade_df
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
bins = [0,584,629,644,675]
labels = ["< $584","$585-$629","$630-$644","$645-$675"]

school_group_df["Per Student Spending Range"] = pd.cut(school_group_df["Budget Per Student"],bins=bins,labels=labels,include_lowest=False)

grouped_by_spending = school_group_df.groupby(["Per Student Spending Range"])

math_avg_by_spend = grouped_by_spending["Average Math Score"].mean()

read_avg_by_spend = grouped_by_spending["Average Reading Score"].mean()

passing_math_by_spend = grouped_by_spending["% Passing Math"].mean()

passing_read_by_spend = grouped_by_spending["% Passing Reading"].mean()

overall_pass_by_spend = grouped_by_spending["% Passing Overall"].mean()

grouped_by_spending_df = pd.DataFrame({
    "Average Math Score":math_avg_by_spend,
    "Average Reading Score":read_avg_by_spend,
    "% Passing Math":passing_math_by_spend,
    "% Passing Reading":passing_read_by_spend,
    "% Passing Overall":overall_pass_by_spend
})

grouped_by_spending_df.head()
```
